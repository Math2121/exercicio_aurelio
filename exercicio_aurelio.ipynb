```python
import pandas as pd 
import numpy as np  
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
```

```python

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
new_names = ['altura_sépala','largula_sépala','altura_pétala','largura_pétala','tipo_de_iris']
dataset = pd.read_csv(url, names=new_names, skiprows=0, delimiter=',')
```

```python
saida_esperada = dataset['tipo_de_iris']

dados = dataset.drop(['tipo_de_iris'], axis=1)

saida_esperada = pd.get_dummies(saida_esperada)
```

```python
dados_treino, dados_teste, saida_esperada_treinamento, saida_esperada_teste = train_test_split(dados,saida_esperada, test_size=0.15) 

dados_treino = np.array(dados_treino).astype(np.float32)
dados_teste  = np.array(dados_teste).astype(np.float32)
saida_esperada_treinamento = np.array(saida_esperada_treinamento).astype(np.float32)
saida_esperada_teste  = np.array(saida_esperada_teste).astype(np.float32)
```

```python


Model = MLPClassifier(hidden_layer_sizes=(3,2), max_iter=10000, alpha=0.15, 
                     solver='sgd', verbose=1,  random_state=233) 

modelo_treino = Model.fit(dados_treino,saida_esperada_treinamento)
saida_predicao = Model.predict(dados_teste)
```

```python

Iris_setosa = np.array( [[0,1,0]] )
Iris_versicolor = np.array( [[0,0,0]] )
Iris_virginica  = np.array( [[0,0,1]] )


entrada_exemplo = np.array([[5.9, 3. , 5.1, 1.8]])

y_previsto = Model.predict(entrada_exemplo)

# Caso a saída seja Iris setosa (100):
if np.array_equal(y_previsto, Iris_setosa) == True:
  print('Iris setosa')

# Caso a saída seja Iris versicolor (010):
if np.array_equal(y_previsto, Iris_versicolor) == True:
  print('Iris versicolor')

# Caso a saída seja Iris virginica (001):
if np.array_equal(y_previsto, Iris_virginica) == True:
  print('Iris virginica')
```
